```python
def iqr_bounds(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return lower, upper

outliers= iqr_bounds(df['SepalLengthCm'])
print("SepalLengthCm outlier bounds:", outliers)
```

```python
#1. What is the average sepal length of all flowers?

df['SepalLengthCm'].mean()
```

```python
#2. Which flower species has the widest petals on average?

df.groupby('Species')['PetalWidthCm'].mean()
```

```python
#3. What is the range (min, max) of petal length across species?

df.groupby('Species')['PetalLengthCm'].agg(['min', 'max'])

#plotting boxplot for petal length across species
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Species', y='PetalLengthCm', palette='pastel')
plt.figure(figsize=(8,6))
plt.show()
```

```python
#4. Which species has the highest variance in sepal width?

df.groupby('Species')['SepalWidthCm'].var()
```

```python
#5. Is the mean petal length of Setosa significantly different from Versicolor? (→ t-test)

from scipy import stats
setosa_petal_length = df[df['Species'] == 'Iris-setosa']['PetalLengthCm']
versicolor_petal_length = df[df['Species'] == 'Iris-versicolor']['PetalLengthCm']
t_stat, p_value = stats.ttest_ind(setosa_petal_length, versicolor_petal_length)
t_stat, p_value
```

```python
#6. Do Setosa and Virginica flowers have the same average sepal width?

setosa_sepal_width = df[df['Species'] == 'Iris-setosa']['SepalWidthCm']
virginica_sepal_width = df[df['Species'] == 'Iris-virginica']['SepalWidthCm']
t_stat, p_value = stats.ttest_ind(setosa_sepal_width, virginica_sepal_width)
t_stat, p_value
```

```python
#7. Is there a significant difference in sepal length between Versicolor and Virginica?

versicolor_sepal_length = df[df['Species'] == 'Iris-versicolor']['SepalLengthCm']
virginica_sepal_length = df[df['Species'] == 'Iris-virginica']['SepalLengthCm']
t_stat, p_value = stats.ttest_ind(versicolor_sepal_length, virginica_sepal_length)
t_stat, p_value
```

```python
#8. Is there a correlation between sepal length and petal length?

df[['SepalLengthCm', 'PetalLengthCm']].corr()

#plotting the correlation
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='SepalLengthCm', y='PetalLengthCm', hue='Species', palette='deep')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.title('Scatter Plot: Sepal Length vs Petal Length by Species')
plt.legend(title='Species')
plt.show()
```

```python
#9. Does sepal width correlate with petal width?

df[['SepalWidthCm', 'PetalWidthCm']].corr()

#ploting the correlation
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='SepalWidthCm', y='PetalWidthCm', hue='Species', palette='deep')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Scatter Plot: Sepal Width vs Petal Width by Species')
plt.show()

```
